```python
import pandas as pd
import tensorflow as tf
from numpy import loadtxt
from keras.models import load_model
from keras.callbacks import EarlyStopping
```

```python
file = open('dataset/sequences_matrix.csv', 'rb')
sequences_matrix = loadtxt(file,delimiter = ",")
print(sequences_matrix)
```

```python
y_train = pd.read_csv('dataset/y_train.csv', index_col = 0, squeeze = True)
print(y_train)
```

```python
model = tf.keras.models.load_model('model/lstm_model.h5')
```

```python
callbacks = EarlyStopping(monitor='val_loss',min_delta=0.0001,patience=3)
```

```python
history = model.fit(sequences_matrix,y_train,batch_size=128,epochs=50,
          validation_split=0.2,callbacks=[callbacks])
```

```python
history = model.fit(sequences_matrix,y_train,batch_size=64,epochs=100,
          validation_split=0.2,callbacks=[callbacks])
```
