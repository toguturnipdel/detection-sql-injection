```python
# Read X_train with Series Type
X_train = pd.read_csv('dataset/X_train.csv', index_col = 0, squeeze = True)
print(X_train)
```

```python
max_words = 24098
max_len = 150
tok = Tokenizer(num_words=max_words)
tok.fit_on_texts(X_train)
```

```python
# Mengubah tiap huruf menjadi urutan bilangan bulat sesuai dengan kamus word_index
def process_data(input_val):
  sample_texts_list = []
  # Change str to list type
  sample_texts_list.append(input_val)
  txts = tok.texts_to_sequences(sample_texts_list)
  input_val = sequence.pad_sequences(txts, maxlen=max_len)
  return input_val
  
def predict_sqli_attack():
  # Untuk Pengulangan Masukan Inputan
  repeat=True
  
  # Style pada Console
  space=''
  for i in range(20):
    space+= "-"
    
  print(space) 
  # Inputan yang ingin diprediksi
  input_val=input("Masukkan Link yang ingin di deteksi: ")
  print(space)
  
  # Jika Inputan 0 Loopingan Selesai
  if input_val== '0':
    repeat=False
  
  result=loaded_model.predict(process_data(input_val))
  print(space)
  if repeat == True:
    if result > 0.5:
      print("Hati\" !Inputan Merupakan SQL Injection")
      print(space)
    elif result <= 0.5:
      print("Inputan Aman")
      print(space)
    predict_sqli_attack()
  elif repeat == False:
    print("Sampai Jumpa")
```
